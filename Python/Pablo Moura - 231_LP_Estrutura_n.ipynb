### 3. Elabore uma estrutura para representar um produto (código, nome, preço). **Crie uma função** para cadastrar 5 produtos. **Crie outra função** para aplicar 10% de aumento no preço do produto e apresente, por meio **de outra função**, todos os dados do produtos cadastrados, após o aumento. Construa uma função para cada opção do menu a seguir:
Menu do Sistema
1. Cadastrar
2. Reajustar
3. Visualizar
4. Sair

Qual opção deseja?

```
class TipoProduto:
    codigo = 0
    nome = ''
    preco = 0.0

def menu():
    print('\nMenu do Sistema')
    print('1. Cadastrar')
    print('2. Reajustar')
    print('3. Visualizar')
    print('4. Consulta por nome')
    print('5. Alterar por código')
    print('6. Sair')
    opcao = int(input('Qual opção deseja? '))
    return opcao

def cadastrar(vet_prod):# vet_prod parâmetro que entra com valor vazio
    for i in range(3):
        p = TipoProduto()# inicializando a vairável p
        p.codigo = int(input('Informe o código: '))
        p.nome = input('Informe o nome: ')
        p.preco = float(input('Informe o preço: '))
        vet_prod.append(p)
    return vet_prod # sai o vet_prod preenchido/ALTERADO com 3 produtos

def visualizar(vp): #o vetor vp entra com conteúdo
# mas não há necessidade de ter retorno, pois NÃO ALTERA o vetor
    if len(vp) > 0:
        for i in range(len(vp)):
            print(f'Código: {vp[i].codigo} \tNome: {vp[i].nome} \tPreço: {vp[i].preco:.2f}')
    else:
        print('Não há produto cadastrado!')

def reajustar(vet_prod):#o vetor entra sem aumento
    if len(vet_prod) > 0:
        for i in range(len(vet_prod)):
            vet_prod[i].preco += vet_prod[i].preco * 10 / 100
        print('Reajuste realizado com sucesso!')
    else:
        print('Não há produto cadastrado!')
    return vet_prod # o vetor sai com alteração de preço

def consulta_nome(vet_prod):
    if len(vet_prod) > 0: # verificando se tem elementos no vetor
        nome_pesquisa = input('Qual nome que encontrar? ')# produto externo
        for prod in vet_prod:
            if nome_pesquisa.lower() in prod.nome.lower(): # upper()
                print(f'Código: {prod.codigo} \tPreço: {prod.preco:.2f}')     
        else:
            print('Não há produto cadastrado!')
    else:
        print('Não há produto cadastrado!')

def alterar(vet_prod):
    if len(vet_prod) > 0:
        codigo_pesquisa = int(input('Qual código deseja pesquisar? '))
        for i in range(len(vet_prod)):
            if vet_prod[i].codigo == codigo_pesquisa:
                #vou alterar o nome e o preço
                vet_prod[i].nome = input('Digite o nome a alterar: ')
                vet_prod[i].preco = float(input('Digite o preço a alterar: '))
    else:
        print('Não há produto cadastrado!')
    return vet_prod # retorna porque o nome e o preço foram alterados

def excluir(vet_prod):
    if len(vet_prod) > 0:
        codigo_pesquisa = int(input('Qual código deseja pesquisar? '))
        for i in range(len(vet_prod)):
            if vet_prod[i].codigo == codigo_pesquisa:
                indice = i
        vet_prod.pop(indice)
        print('Exclusão com sucesso!')

    else:
        print('Produto não encontrado')
    
    return vet_prod

def main():
    op = menu()
    vet_produto = []
    while op >= 1 and op <= 5:
        if op == 1:
            vet_produto = cadastrar(vet_produto)#vet_produto entra vazio
        elif op == 2:
            vet_produto = reajustar(vet_produto)
        elif op == 3:
            visualizar(vet_produto)
        elif op == 4:
            consulta_nome(vet_produto)
        elif op == 5:
            vet_produto = alterar(vet_produto)
        elif op == 6:
            vet_produto = excluir(vet_produto)
        op = menu()

main()
```

### 4. Uma escola precisa montar o cadastro geral de seus alunos. Este cadastro deverá conter as seguintes informações por aluno: nome completo, data de nascimento, telefone, endereço e série atual. Levando em conta que esta escola possui no máximo 500 alunos, como você faria para estruturar estas informações num sistema de gerenciamento para a escola? Implemente utilizando estrutura. Também **use a criação de funções para cada operação**. Use o menu a seguir:
Menu de opções:
1. Cadastrar alunos
2. Consulta por nome
3. Visualizar todos os dados

```
# Digite seu código aqui
class TipoAluno:
    nome = ''
    nascimento = ''
    fone = 0
    endereco = ''
    serie = 0

def menu():
    print('\nMenu do Sistema')
    print('1. Cadastrar aluno')
    print('2. Consulta por nome')
    print('3. Visualizar todos os dados')
    print('4. Sair')
    opcao = int(input('Qual opção deseja? '))
    return opcao

def cadastrar(v_aluno):
    if len(v_aluno) < 500:
        a = TipoAluno()
        a.nome = input('Informe o nome do aluno: ')
        a.nascimento = input('Informe o nascimento do aluno: ')
        a.fone = int(input('Informe o telefone do aluno: '))
        a.endereco = input('Informe o endereço do aluno: ')
        a.serie = int(input('Informe a série do aluno: '))
        v_aluno.append(a)
        print('Aluno cadastrado com sucesso!')
    else:
        print('Limite de alunos excedido!')
    return v_aluno

def consulta_nome(v_aluno):
    if len(v_aluno) > 0: # verificando se tem elementos no vetor
        nome_pesquisa = input('Qual nome quer encontrar? ')
        for estudante in v_aluno:
            if nome_pesquisa.lower() in estudante.nome.lower(): # upper()
                print(f'Nascimento: {estudante.nascimento} \tFone: {estudante.fone} \tEndereco: {estudante.endereco} \tSérie: {estudante.serie}')     
    else:
        print('Não há aluno cadastrado!')

def visualizar(v_aluno):
    if len(v_aluno) > 0:
        for i in range(len(v_aluno)):
            print(f'Nome: {v_aluno[i].nome} \tNascimento: {v_aluno[i].nascimento} \tFone: {v_aluno[i].fone} \tEndereco: {v_aluno[i].endereco} \tSérie: {v_aluno[i].serie}')
    else:
        print('Não há alunos cadastrados!')

def main():
    op = menu()
    vet_aluno = []
    while op >=1 and op <= 3:
        if op == 1:
            vet_aluno = cadastrar(vet_aluno)
        elif op == 2:
            consulta_nome(vet_aluno)
        elif op == 3:
            visualizar(vet_aluno)
        op = menu()

main()
```

### 5. Faça um programa que realize o cadastro de contas bancarias com as seguintes informações: numero da conta, nome do titular e saldo. O banco permitirá o cadastramento de 15 contas. **Crie uma função para cada opção do menu** a seguir. Utilize a estrutura na passagem de parâmetro:
Menu de opções:
1. Cadastrar contas
2. Visualizar todas as contas
3. Consultar por nome
4. Alterar nome e/ou saldo de um número de conta
5. Excluir a conta com menor saldo
6. Sair

Observação: 

* No item de menu 4. Alterar nome e/ou saldo de um número de conta, entenda que apenas pode ser alterado o nome e o saldo depois que você pesquisar pelo número da conta.
* No item 5 do menu, encontre o menor saldo dentre todos cadastrados, **depois exclua esta ÚNICA conta.**. O assunto de excluir algo de um vetor foi dado na disciplina de Algoritmo.



```
# Percorra todo os vetor e quando a condição de encontrar um menor saldo for executada, guarde o índice
# Fora do laço de repetição, exclua pelo índice, por exemplo com o comando pop(índice)

class TipoConta:
    numero = 0
    titular = ''
    saldo = 0.0

def menu():
    print('\nMenu do Sistema')
    print('1. Cadastrar contas')
    print('2. Visualizar todas as contas')
    print('3. Consultar por nome')
    print('4. Alterar nome e/ou saldo de um número de conta')
    print('5. Excluir a conta com menor saldo')
    print('6. Sair')
    opcao = int(input('Qual opção deseja? '))
    return opcao

def cadastrar(conta):
    if len(conta) <= 15:
        c = TipoConta()
        c.numero = int(input('Cadastrar numero de conta: '))
        c.titular = input('Informe o nome do titular da conta: ')
        c.saldo = float(input('Quanto deseja depositar? : '))
        conta.append(c)
        print('Cliente cadsatrado com sucesso!')
    else:
        print('Limite de clientes excedido')
    return conta

def visualizar(conta):
    if len(conta) > 0:
        for i in range(len(conta)):
           print(f'Número da conta: {conta[i].numero} \tTitular: {conta[i].titular} \tSaldo: {conta[i].saldo}')

def consulta(conta):
    if len(conta) > 0:
        pesquisa = input('Qual nome deseja encontrar? ')
        for n in conta:
            if pesquisa.lower() in n.titular.lower():
                print(f'Conta: {n.numero} \tSaldo: {n.saldo}')

def alterar(conta):
    if len(conta) > 0:
        codigo = int(input('Qual número de conta deseja pesquisar? '))
        for i in range(len(conta)):
            if conta[i].numero == codigo:
                conta[i].titular = input('Digite o nome a alterar: ')
                deposito = float(input('Quanto deseja depositar/retirar? '))
                conta[i].saldo += deposito
                print(f'O novo saldo é: {conta[i].saldo}')
    else:
        print('Não há produto cadastrado!')
    return conta

def excluir(conta):
    if len(conta) > 0:
        menor_saldo = conta[0].saldo
        indice_menor_saldo = 0
        for i in range(1, len(conta)):
            if conta[i].saldo < menor_saldo:
                menor_saldo = conta[i].saldo
                indice_menor_saldo = i
        conta.pop(indice_menor_saldo)
        print(f'A conta foi excluída com sucesso.')
    else:
        print('Não há contas cadastradas.')

    return conta


def main():
    op = menu()
    vet_conta = []
    while op >=1 and op <= 5:
        if op == 1:
            vet_conta = cadastrar(vet_conta)
        elif op == 2:
            visualizar(vet_conta)
        elif op == 3:
            consulta(vet_conta)
        elif op == 4:
            vet_conta = alterar(vet_conta)
        elif op == 5:
            vet_conta = excluir(vet_conta)
        op = menu()

main()
```

### 6. Elabore uma estrutura para representar um Funcionario (código, nome, endereço, salário). Construa uma função para cada opçao do menu a seguir:

Menu de opções:
1. Cadastrar funcionários

```
class TipoFuncionario:
    codigo = 0
    nome = ''
    endereco = None
    salario = 0.0

class TipoEndereco:
    logradouro = ''
    numero = 0
    bairro = ''
    cidade = ''

def menu():
    print('\nMenu do Sistema')
    print('1. Cadastrar funcionários')
    print('2. Visualizar todos os dados')
    print('3. Sair')
    opcao = int(input('Qual opção deseja? '))
    return opcao

def cadastrar():
    func = TipoFuncionario()
    func.codigo = int(input('Informe o código do funcionário: '))
    func.nome = input('Informe o nome do funcionário: ')
    func.salario = float(input('Informe o salário do funcionário: '))

    endereco = TipoEndereco()
    endereco.logradouro = input('Informe o logradouro: ')
    endereco.numero = int(input('Informe o número: '))
    endereco.bairro = input('Informe o bairro: ')
    endereco.cidade = input('Informe a cidade: ')

    func.endereco = endereco
    return func

def visualizar(funcionarios):
    for func in funcionarios:
        print('Código:', func.codigo)
        print('Nome:', func.nome)
        print('Salário:', func.salario)
        print('Endereço:')
        print('Logradouro:', func.endereco.logradouro)
        print('Número:', func.endereco.numero)
        print('Bairro:', func.endereco.bairro)
        print('Cidade:', func.endereco.cidade)
        print('--------------------------')

def main():
    op = menu()
    funcionarios = []
    
    while op >= 1 and op <= 3:
        if op == 1:
            func = cadastrar()
            funcionarios.append(func)
            print('Funcionário cadastrado com sucesso!')
        elif op == 2:
            if len(funcionarios) > 0:
                visualizar(funcionarios)
            else:
                print('Nenhum funcionário cadastrado.')
        elif op == 3:
            print('Saindo do sistema...')
            break
        else:
            print('Opção inválida!')

        op = menu()

main()
```

### 7. Elabore uma estrutura para representar um Produto (código, nome, data_fabricacao, data_validade, preço). Para o membro data_fabricacao e data_validade, deve-se criar outra estrutura Data (dia, mês, ano). **Utilize aninhamento de estruturas** para resolver este desenvolvimento. Construa uma função para cada opçao do menu a seguir:

Menu de opções:
1. Cadastrar produtos
2. Visualizar todos os dados

```
# Digite seu código aqui
class Data:
    dia = 0
    mes = 0
    ano = 0

class Produto:
    codigo = 0
    nome = ''
    data_fabricacao = None
    data_validade = None
    preco = 0.0

def menu():
    print('\nMenu do Sistema')
    print('1. Cadastrar produtos')
    print('2. Visualizar todos os dados')
    print('3. Sair')
    opcao = int(input('Qual opção deseja? '))
    return opcao

def cadastrar():
    prod = Produto()
    prod.codigo = int(input('Informe o código do produto: '))
    prod.nome = input('Informe o nome do produto: ')
    prod.preco = float(input('Informe o preço do produto: '))

    fabricacao = Data()
    fabricacao.dia = int(input('Informe o dia de fabricação: '))
    fabricacao.mes = int(input('Informe o mês de fabricação: '))
    fabricacao.ano = int(input('Informe o ano de fabricação: '))

    validade = Data()
    validade.dia = int(input('Informe o dia de validade: '))
    validade.mes = int(input('Informe o mês de validade: '))
    validade.ano = int(input('Informe o ano de validade: '))

    prod.data_fabricacao = fabricacao
    prod.data_validade = validade
    return prod

def visualizar(produtos):
    for prod in produtos:
        print('Código:', prod.codigo)
        print('Nome:', prod.nome)
        print('Preço:', prod.preco)
        print('Data de Fabricação:', prod.data_fabricacao.dia, '/', prod.data_fabricacao.mes, '/', prod.data_fabricacao.ano)
        print('Data de Validade:', prod.data_validade.dia, '/', prod.data_validade.mes, '/', prod.data_validade.ano)
        print('--------------------------')

def main():
    op = menu()
    produtos = []
    
    while op >= 1 and op <= 3:
        if op == 1:
            prod = cadastrar()
            produtos.append(prod)
            print('Produto cadastrado com sucesso!')
        elif op == 2:
            if len(produtos) > 0:
                visualizar(produtos)
            else:
                print('Nenhum produto cadastrado.')
        elif op == 3:
            print('Saindo do sistema...')
            break
        else:
            print('Opção inválida!')

        op = menu()

main()
```

### 8. Elabore duas estruturas, como é apresentado a seguir:

CLIENTE |  DOCUMENTOS |
 -------|--------------
cod_cli |   num_doc   |
nome    |   cod_cli   |
fone    |   dia_venc  |
        |   dia_pag   |
        |   valor     |
        |   juros     |

* Sabe-se que um documento só pode ser cadastrado para um cliente que já exista. 
* Considere que podem existir, no máximo, 15 clientes e 30 documentos. Crie um vetor para clientes e outro para documentos. 

* Crie um menu para a realização de cada uma das operações especificadas a seguir:

**** SISTEMA GERENCIADOR DE CLIENTES E DOCUMENTOS ****

1. Cadastrar clientes

2. Relatório de clientes

3. Cadastrar documentos 

4. Relatório de documentos

5. Excluir clientes sem documentos

6. Excluir documentos individuais pelo número

7. Excluir documentos por cliente

8. Excluir documentos por período 

9. Alterar as informações dos clientes 

10. Mostrar o total de documentos de determinado cliente

11. Sair

Qual opção deseja?

.................................................................................................

Para cada item do menu, desenvolva uma função.

A seguir são apresentados os detalhes de implementação de cada opção do menu:

1. Cadastrar clientes — não pode existir mais que um cliente com o mesmo código.

2. Relatório de clientes - listar todos os clientes cadastrados.

3. Cadastrar documentos — ao cadastrar um documento, **se o dia de pagamento for maior que o dia de vencimento, calcular o campo ‘juros’ do registro documentos (5% sobre o valor original do documento)**.

4. Relatório de documentos - listar todos os documentos cadastrados.

5. Excluir clientes — um cliente só poderá ser excluído se não existir nenhum documento associado a ele.

6. Excluir documentos individuais — por meio de seu número. Caso o documento não exista, o programa deverá mostrar a mensagem "Documento não encontrado".

7. Excluir documentos por cliente — o programa deverá informar o código do cliente e excluir todos os seus documentos. Caso o cliente não exista, deverá mostrar a mensagem "Cliente não encontrado".

8. Excluir documentos por período — o programa deverá informar o dia inicial e o dia final e excluir todos os documentos que possuam data de vencimento nesse período.

9. Alterar as informações sobre os clientes — **só NÃO altere o código do cliente**.

10. Mostrar o total de documentos de determinado cliente.

```
# Por onde começo a programar esse exercício?
# Sempre comece da parte mais simples/talvez a menor. 
# Vai implementando e testando.

# Digite seu código aqui
class Cliente:
    cod_cli = 0
    nome = ''
    fone = ''

class Documento:
    num_doc = ''
    cod_cli = 0
    dia_venc = 0
    dia_pag = 0
    valor = 0.0
    juros = 0.0

def menu():
    print('\n** SISTEMA GERENCIADOR DE CLIENTES E DOCUMENTOS **')
    print('1. Cadastrar clientes')
    print('2. Relatório de clientes')
    print('3. Cadastrar documentos')
    print('4. Relatório de documentos')
    print('5. Excluir clientes sem documentos')
    print('6. Excluir documentos individuais pelo número')
    print('7. Excluir documentos por cliente')
    print('8. Excluir documentos por período')
    print('9. Alterar informações dos clientes')
    print('10. Mostrar total de documentos de um cliente')
    print('11. Sair')
    opcao = int(input('Qual opção deseja? '))
    return opcao

def cadastrar_clientes(clientes):
    if len(clientes) >= 15:
        print('Limite máximo de clientes atingido.')
        return
    
    cliente = Cliente()
    cliente.cod_cli = int(input('Informe o código do cliente: '))
    for c in clientes:
        if c.cod_cli == cliente.cod_cli:
            print('Cliente com código já existente.')
            return
    
    cliente.nome = input('Informe o nome do cliente: ')
    cliente.fone = input('Informe o telefone do cliente: ')
    
    clientes.append(cliente)
    print('Cliente cadastrado com sucesso!')

def relatorio_clientes(clientes):
    print('--- Relatório de Clientes ---')
    if len(clientes) == 0:
        print('Nenhum cliente cadastrado.')
    else:
        for cliente in clientes:
            print('Código:', cliente.cod_cli)
            print('Nome:', cliente.nome)
            print('Telefone:', cliente.fone)
            print('--------------------------')

def cadastrar_documentos(documentos, clientes):
    if len(documentos) >= 30:
        print('Limite máximo de documentos atingido.')
        return
    
    documento = Documento()
    documento.num_doc = input('Informe o número do documento: ')
    for doc in documentos:
        if doc.num_doc == documento.num_doc:
            print('Documento com número já existente.')
            return
    
    documento.cod_cli = int(input('Informe o código do cliente: '))
    found = False
    for cliente in clientes:
        if cliente.cod_cli == documento.cod_cli:
            found = True
            break
    if not found:
        print('Cliente não encontrado.')
        return
    
    documento.dia_venc = int(input('Informe o dia de vencimento: '))
    documento.dia_pag = int(input('Informe o dia de pagamento: '))
    documento.valor = float(input('Informe o valor do documento: '))
    if documento.dia_pag > documento.dia_venc:
        documento.juros = documento.valor * 0.05
    
    documentos.append(documento)
    print('Documento cadastrado com sucesso!')

def relatorio_documentos(documentos):
    print('--- Relatório de Documentos ---')
    if len(documentos) == 0:
        print('Nenhum documento cadastrado.')
    else:
        for documento in documentos:
            print('Número:', documento.num_doc)
            print('Código do Cliente:', documento.cod_cli)
            print('Dia de Vencimento:', documento.dia_venc)
            print('Dia de Pagamento:', documento.dia_pag)
            print('Valor:', documento.valor)
            print('Juros:', documento.juros)
            print('--------------------------')

def excluir_clientes_sem_documentos(clientes, documentos):
    clientes_excluidos = []
    for cliente in clientes:
        has_documentos = False
        for documento in documentos:
            if documento.cod_cli == cliente.cod_cli:
                has_documentos = True
                break
        if not has_documentos:
            clientes_excluidos.append(cliente)
    
    if len(clientes_excluidos) == 0:
        print('Não há clientes sem documentos.')
    else:
        for cliente in clientes_excluidos:
            clientes.remove(cliente)
        print('Clientes excluídos sem documentos:', len(clientes_excluidos))

def excluir_documento_individual(documentos, numero):
    documento_encontrado = False
    for documento in documentos:
        if documento.num_doc == numero:
            documentos.remove(documento)
            documento_encontrado = True
            break
    
    if documento_encontrado:
        print('Documento excluído com sucesso.')
    else:
        print('Documento não encontrado.')

def excluir_documentos_por_cliente(documentos, clientes, cod_cli):
    cliente_encontrado = False
    documentos_excluidos = []
    for documento in documentos:
        if documento.cod_cli == cod_cli:
            documentos_excluidos.append(documento)
    for documento in documentos_excluidos:
        documentos.remove(documento)
    if len(documentos_excluidos) > 0:
        cliente_encontrado = True
    
    if cliente_encontrado:
        print('Documentos excluídos do cliente com código', cod_cli, ':', len(documentos_excluidos))
    else:
        print('Cliente não encontrado.')

def excluir_documentos_por_periodo(documentos, dia_inicial, dia_final):
    documentos_excluidos = []
    for documento in documentos:
        if documento.dia_venc >= dia_inicial and documento.dia_venc <= dia_final:
            documentos_excluidos.append(documento)
    for documento in documentos_excluidos:
        documentos.remove(documento)
    
    print('Documentos excluídos no período de', dia_inicial, 'a', dia_final, ':', len(documentos_excluidos))

def alterar_informacoes_clientes(clientes):
    cod_cli = int(input('Informe o código do cliente que deseja alterar: '))
    for cliente in clientes:
        if cliente.cod_cli == cod_cli:
            cliente.nome = input('Informe o novo nome do cliente: ')
            cliente.fone = input('Informe o novo telefone do cliente: ')
            print('Informações do cliente atualizadas.')
            return
    
    print('Cliente não encontrado.')

def total_documentos_cliente(documentos, cod_cli):
    total_documentos = 0
    for documento in documentos:
        if documento.cod_cli == cod_cli:
            total_documentos += 1
    
    print('Total de documentos do cliente', cod_cli, ':', total_documentos)

def main():
    clientes = []
    documentos = []
    
    while True:
        opcao = menu()
        
        if opcao == 1:
            cadastrar_clientes(clientes)
        elif opcao == 2:
            relatorio_clientes(clientes)
        elif opcao == 3:
            cadastrar_documentos(documentos, clientes)
        elif opcao == 4:
            relatorio_documentos(documentos)
        elif opcao == 5:
            excluir_clientes_sem_documentos(clientes, documentos)
        elif opcao == 6:
            numero = input('Informe o número do documento que deseja excluir: ')
            excluir_documento_individual(documentos, numero)
        elif opcao == 7:
            cod_cli = int(input('Informe o código do cliente para excluir seus documentos: '))
            excluir_documentos_por_cliente(documentos, clientes, cod_cli)
        elif opcao == 8:
            dia_inicial = int(input('Informe o dia inicial do período: '))
            dia_final = int(input('Informe o dia final do período: '))
            excluir_documentos_por_periodo(documentos, dia_inicial, dia_final)
        elif opcao == 9:
            alterar_informacoes_clientes(clientes)
        elif opcao == 10:
            cod_cli = int(input('Informe o código do cliente para mostrar o total de documentos: '))
            total_documentos_cliente(documentos, cod_cli)
        elif opcao == 11:
            print('Saindo do sistema...')
            break
        else:
            print('Opção inválida! Tente novamente.')

main()
```
