###7. (Função sem retorno com parâmetro) Faça uma função/método para calcular a tabuada de um ***intervalo*** informado pelo usuário, por exemplo, tabuada do 3 ao 8.

```
# Digite seu código aqui.
def tabuada(x,y,z):
    for i in range(y,z+1):
        print(x,'X',i,'=',x * i)

def main():
    valor1 = int(input('digite o valor multiplicado: '))
    valor2 = int(input('digite o primeiro numero do intervalo: '))
    valor3 = int(input('digite o segundo numero do intervalo: '))
    tabuada(valor1, valor2, valor3)

main()
```

### 11. (Função sem retorno com parâmetro) Faça uma função/método para verificar e contar quantas letras **a** tem numa frase. Não use NENHUMA função pronta da linguagem Python.

```
# Digite seu código aqui.
def resultado(a):
    soma = 0
    for i in range(len(a)):
        if a[i] == 'a' or a[i] == 'A':
            soma += 1
    print('Tem',soma,'letras "A(s)" na frase')

def main():
    frase = input('digite uma frase aqui: ')
    resultado(frase)

main()
```

### 12. Crie outra função que fará a inserção de mais um filho para este funcionário no vetor, apresente após a verificação a relação de nomes de funcionários e a quantidade de filhos de todos.

```
# Digite seu código aqui.
def cadastro(a,b):
    verificar = []
    soma = 0
    for j in range(len(a)):
        verificar.append(int(input(f'{a[j]} houve quantos filhos nos ultimos 5 anos? ')))
    print(verificar)

    insercao(a,b,verificar)

def insercao(x,y,z):
    soma = []
    for k in range(len(x)):
        soma.append(y[k] + z[k])
        print(f'a quantidade de filhos de {x[k]} é {soma[k]}')
    print(soma)

def main():
    funcionarios = []
    filhos = []
    for i in range(4):
        funcionarios.append(input('nome do funcionário: '))
        filhos.append(int(input('quantidade de filhos ')))
    print(funcionarios, filhos)
    cadastro(funcionarios, filhos)

main()
```
